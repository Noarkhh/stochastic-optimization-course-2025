```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(2137)
m = 100
features = 1
```

```python
x = np.random.randint(m, size=(m, features))
y = (x * 2 + 5 + np.random.normal(0, 5, size=(m, features)))
x_b = np.hstack((x, np.ones((m, 1))))
```

```python
plt.scatter(x, y)
```

```python
learning_rate = 0.0001
weights = np.array([[0.0], [0.0]])
epochs = 100000
losses = []
for _ in range(epochs):
    y_pred = x_b.dot(weights)
    loss = np.sum(np.square(y_pred - y)) / m
    losses.append(loss)
    gradient = (2 / m) * x_b.T.dot(y_pred - y)
    weights -= learning_rate * gradient
```

```python
plt.plot(x, x_b.dot(weights), color="purple", label="result")
plt.plot(x, x_b.dot(np.array([[2.0], [3.0]])), color="red", label="truth")
plt.scatter(x, y)
plt.title("Gradient descent result and ground truth comparison")
plt.legend()
```

```python
plt.plot(losses[4:])
plt.ylabel("loss")
plt.xlabel("epoch")
plt.title("Loss during training")
```

```python
def mse(w1, w2):
    weights = np.array([[w1], [w2]])
    y_pred = x_b.dot(weights)
    loss = np.sum(np.square(y_pred - y)) / m
    return loss

A = np.linspace(-1, 4, 100)
B = np.linspace(-30, 30, 100)
Y = np.array([[np.log10(mse(a, b)) for a in A] for b in B])
```

```python
plt.contourf(A, B, Y)
plt.plot(0, 0, label="start weights", marker="o", color="red")
plt.plot(weights[0][0], weights[1][0], label="end weights", marker="o", color="orange")
plt.legend()
plt.title("Loss landscape")
plt.ylabel("b")
plt.xlabel("a")
```

```python
best_weights = (np.linalg.inv(x_b.T @ x_b) @ x_b.T).dot(y)
print(f"analytical MSE: {mse(best_weights[0][0], best_weights[1][0])}")
print(f"gradient descent MSE: {mse(weights[0][0], weights[1][0])}")

plt.plot(x, x_b.dot(weights), color="purple", label="result")
plt.plot(x, x_b.dot(best_weights), color="red", label="analytical")
plt.scatter(x, y)
plt.title("Gradient descent result and analytical solution comparison")
plt.legend()
```
