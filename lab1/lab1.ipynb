```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(2137)
```

```python
x = np.random.randint(100, size=(100,))
y = (x * 2 + 5 + np.random.normal(0, 5, size=(100,))).reshape((-1, 1))
```

```python
plt.scatter(x, y)
```

```python
x = np.expand_dims(x, 1)
new_x = np.hstack((x, np.ones((100, 1))))
```

```python
learning_rate = 0.00001
weights = np.array([[2.0], [3.0]])
for _ in range(500000):
    y_pred = new_x.dot(weights)
    gradient = (2 / 100) * new_x.T.dot(y_pred - y)
    weights -= learning_rate * gradient
```

```python
weights
```

```python
plt.plot(x, new_x.dot(weights), color="green", label="Initial Line (No GD)")
plt.scatter(x, y)
```
